# Part 1: Twitter Network Analysis
### Exercise 1 description 
Build the network of retweets. We will now build a network that has as nodes the Twitter handles of the members of the house, and direct edges between nodes A and B if A has retweeted content posted by B. We will build a weighted network, where the weight of an edge is equal to the number of retweets. You can build the network following the steps below (and you should be able to reuse many of the functions you wrote in previous weeks):

* Consider the 200 most recent tweets written by each member of the house (use the files here, or the ones you produced in Part 1). For each file, use a regular expression to find retweets and to extract the Twitter handle of the user whose content was retweeted. All retweets begin with "RT @originalAuthor:", where "originalAuthor" is the handle of the user whose content was retweeted (and the part of the text you want to extract).

* For each retweet, check if the handle retweeted is the one of a member of the house. If yes, keep it. If no, discard it.

* Use a NetworkX DiGraph to store the network. Use weighted edges to account for multiple retweets. Store also the party of each member as a node attribute (use the data in this file, or the data you downloaded in Part 1). Remove self-loops (edges that connect a node with itself).



```python
def readFromFile(filename):
    with io.open('./'+filename, encoding='utf-8') as f:
        return f.read()

# We read in the twitter names from the provided file
df = pd.read_csv('./H115_tw.csv')
twitterNames = {}
twitterFiles = {}
for i in range(len(df)):
    tweet_name = df['tw_name'][i]
    party = df['Party'][i]
    twitterNames[tweet_name] = party

# We read in the tweets for each representative
for name in twitterNames.keys():
    f = readFromFile('./tweets/%s' % name)
    twitterFiles[name] = f

# Finding retweets to other house members
congress_retweets = {}
for name, tweets in twitterFiles.items():
    house_retweets = []
    regexp = "RT @([^\s^\:]+)"
    retweets = re.findall(regexp, tweets)
    for handle in retweets:
        if handle in twitterNames:
            house_retweets.append(handle)
    # Every member is linked to a list of retweets of other congress members
    congress_retweets[name] = house_retweets

# Build graph
G = nx.DiGraph()
for name, retweets in congress_retweets.items():
    G.add_node(name, party=twitterNames[name])
    for retweet in retweets:
        # No self-loops
        if (name == retweet):
            continue
        if G.has_edge(name, retweet):
            # If edge already exists, increase weight of it by 1
            G[name][retweet]['weight'] += 1.0
        else:
            G.add_edge(name, retweet, weight=1.0)

```

We then filter the data by removing twitter handles, punctuation and numbers, retweets, etc.

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
def tokenize_twitter(text):
    text = text.encode('ascii','ignore')
    tokens = nltk.word_tokenize(text)

    # Remove handles
    cleanTokens = []
    skip = False
    for i in range(len(tokens)):
        if skip:
            skip = False
            continue
        token = tokens[i]
        # If @ is found, skip this token and the next (which is the handle)
        if token == '@':
            skip = True
        else:
            cleanTokens.append(token)
            skip = False

    # Remove all punctuation and numbers
    regexp = r'[^\W\d]+'
    tempTokens = []
    for token in cleanTokens:
        if re.match(regexp, token):
            tempTokens.append(token)
    cleanTokens = tempTokens[:]
    
    #Remove retweets
    regexp = "(RT)|(.)"
    tempTokens = []
    for token in cleanTokens:
        match = re.match(regexp, token)
        if match.groups()[1]:
            tempTokens.append(token)
    cleanTokens = tempTokens[:]
    
    # Remove all words of length 1 and set lowercase
    cleanTokens = [w.lower() for w in cleanTokens if len(w) > 1]
    stopwords = nltk.corpus.stopwords.words('english')

    # Remove stopwords
    cleanTokens = [w for w in cleanTokens if w not in stopwords]
    
    # Remove tokens starting with http (links)
    cleanTokens = [w for w in cleanTokens if not w.startswith('http')]
    
    return cleanTokens

tokens_demo = tokenize_twitter(democratic)
tokens_rep = tokenize_twitter(republicans)
```

Below we define the functions used in computing the TF-IDF for a list of tokens:

```python
def computeTF(text):
    fdist = nltk.FreqDist(text)
    tf = dict()
    for (i,j) in fdist.most_common():
        tf[i] = float(j) / len(text)
    return tf

def computeIDF(text):
    import math
    idf = dict()
    n = len(text)
    for word in text:
        if word not in idf:
            idf[word] = 1
        else:
            idf[word] += 1
    for word,val in idf.items():
        idf[word] = math.log10(n / float(val))
    return idf

def computeTFIDF(text):
    tf = computeTF(text)
    idf = computeIDF(text)
    tfidf = dict()
    for word,value in tf.items():
        tfidf[word] = value*idf[word]
    return tfidf

tfidfDemo = computeTFIDF(tokens_demo)
tfidfRepub = computeTFIDF(tokens_rep)
```

The TF-IDF is then used to produce the word cloud for each party:

```python
!pip install wordcloud
from wordcloud import WordCloud
def generate_wordcloud(tfidf):
    wordcloud = WordCloud(max_font_size=40).generate_from_frequencies(frequencies=tfidf)
    plt.figure(figsize=(10,10))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()

print("Wordcloud for twitter activity for Democrats")
generate_wordcloud(tfidfDemo)
print("Wordcloud for twitter activity for Republicans")
generate_wordcloud(tfidfRepub)
```

Below we will be loading in the labMT word list and creating the function to calculate the sentiment of a given list of tokens.
1: The LabMT wordlist was created by first finding which words to include, this was done by scanning through Twitter, Google Books (English), music lyrics (1960 to 2007), and the New York Times (1987 to 2007), and picking the most frequent words. Afterwards the word were evaluated by users on Mechanical Turk, on a scale from sad to happy. 

```python
import math
```

```python
def read_file(filename):
    with io.open('./'+filename, encoding='utf-8') as f:
        return f.read()

# Read word list and extract all the various categories
mt_word_list = read_file('Data_Set_S1.txt').splitlines()[3:]
sentiment_dict = {}

# Iterating over every word in list
for row in mt_word_list[1:]:
    cols = row.split("\t")
    word = cols[0]
    # cols[2] is the happiness score avg.
    sentiment_dict[word] = cols[2]

# Function to calculate sentiment given a list of tokens
def get_sentiment(tokens):
    sentiment_sum = 0.0
    count = 0
    for token in tokens:
        if token in sentiment_dict:
            sentiment_sum += float(sentiment_dict[token]) # happiness avg
            count += 1
    # if tweet does not contain any words in the provided list, return None
    if count == 0:
        return None
    avg_hap = sentiment_sum / float(count)
    return avg_hap
```

Below we calculate the  sentiment of all the given tweets

* Create two lists: one including the tweets written by democratic members, and the other including the tweets written by republican members (in the text files, tweets are separated by newlines).

* 2: Calculate the sentiment of each tweet and plot the distribution of sentiment for each of the two lists. Are there significant differences between the two? Which party post more positive tweets?

```python
sentiment_demo = []
sentiment_rep = []
tweets_demo = []
tweets_rep = []
for name, party in twitterNames.items():
    tweets = twitterFiles[name]
    for tweet in tweets.split("\n"):
        tokenized_tweet = tokenize_twitter(tweet)
        sentiment = get_sentiment(tokenized_tweet)
        if not sentiment:
            continue
        if party == 'Republican':
            sentiment_rep.append(sentiment)
            tweets_rep.append(tokenized_tweet)
        else:
            sentiment_demo.append(sentiment)
            tweets_demo.append(tokenized_tweet)
```
